 #### ¿Cuál es el uso más frecuente (Usage_Frequency) de los productos?


```python
df['Usage_Frequency'].value_counts().sort_values(ascending=False)
```

El uso más frecuente de los productos es Occasionaly, seguido por Monthly, sin embargo, me parece extraño

```python
df.loc[df['Usage_Frequency'] == 'Occasional', 'Category'].value_counts().sort_values(ascending=False)
```

Para detallar mejor, decide hacer un conteo de frecuencia de uso por categoria.

Es raro que la mayor cantidad de frecuencia de uso,establecida para los productos sea ocacionalmente
* Observando: como el Serum, es de uso ocacional, cuando suele usarse diario

```python
pd.crosstab(df['Category'], df['Usage_Frequency'])
```

En datos reales esperaria cosas como:

Serum → más Daily/Weekly
Face Mask → más Weekly
Lipstick → más Occasional

Pero aquí:

Todas las categorías tienen números muy parecidos
No hay patrones claros
Todo ronda ~140–180

```python
df['Usage_Frequency'].value_counts(normalize=True)
```

```python
pd.crosstab(df['Category'], df['Usage_Frequency'], normalize='index')
```

No hay relación real entre categoría y frecuencia, entonces
Aunque ‘Occasional’ aparece como la frecuencia más común, el análisis cruzado muestra, que todas las categorías tienen distribuciones muy similares entre frecuencias. 
Esto sugiere que el dataset no refleja patrones reales de uso, sino que probablemente fue generado o balanceado artificialmente.
