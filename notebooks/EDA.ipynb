#### ¿Qué marcas (Brand) tienen más productos?


```python
brandsXproducts =  df.groupby('Brand').size().sort_values(ascending=False)
print(brandsXproducts)
```

Respuesta: La marca con mas productos en el dataset es Milk Makeup, con un total de 426. Las que siguen son Make Up For Ever(414) y Kiehl’s(411) con una diferencia ligeramente menor.

 #### ¿Cuáles son las categorías (Category) más representadas? ¿Hay categorías que aparezcan muy poco?

```python
df['Category'].value_counts().sort_values(ascending=False)
```
