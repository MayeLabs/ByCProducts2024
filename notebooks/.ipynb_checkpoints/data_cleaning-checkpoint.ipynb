#### Transformacion de columnas categoricas

##### *Clasificacion columnas categoricas*

- **Ordinales (jerarquía o clasificacion que implica magnitud o progesion):**
    1. Category: Posee una relacion funcional a traves de una secuencia de uso: Makeup Remover → Foundation → Highlighter.
    2. Usage_Frequency: Tiene un orden claro: Weekly, Occasional , daily.
    3. Product_Size: Tiene una relación de tamaño: 250ml > 100ml > 30ml.
    4. Packaging_Type: No posee un orden implicto, pero puede ser jerarquico en funcionalidad o tamaño.
- **Nominales (Sin orden implicito):**
    1. Brand: Nombre de marcas, sin relacion de orden.
    2. Gender_Target: No tiene orden, no se puede decir que uno es mayor que otro y son independientes no existe una relacion que los conecte.
    3. Main_Ingredient: No tiene jerarquía ni dependencia.
    4. Product_Name: Representan los nombre o identificador
    5. Country_of_Origing: Etiqueta geográfica

Agregar columnas al DF para realizar las transformaciones

##### **Transformacion de columnas categoricas ORDINALES**

##### *Transformacion - Definiendo Orden con un mapeo de las categorías*

```python
# Label Encoding - Maintaining order
# Category: Usage_Frequency
# Order: From lowest to highest frequency

df['Usage_Frequency'].unique()
```

```python
# Dictionary with the order
order = {'Monthly':0,'Occasional':1, 'Weekly':2, 'Daily':3}
# Apply map
df['Usage_Frequency_encoded'] = df['Usage_Frequency'].map(order)
df[['Usage_Frequency', 'Usage_Frequency_encoded']].head()
```

```python
df.head()
```

###### Columna Category

```python
df['Category'].unique()
```

Orden en secuencia de uso:

Makeup Remover → Cleanser → Exfoliator → Face Mask → Face Oil → Serum → Moisturizer → Primer → CC Cream → BB Cream → Foundation → Concealer → Contour → Blush → Highlighter → Bronzer → Eye Shadow → Eyeliner → Mascara → Lip Liner → Lipstick → Lip Gloss → Powder → Setting Spray

```python
#Dictionary
order_category = {'Makeup Remover':0, 'Cleanser':1, 'Exfoliator':2, 'Face Mask':3, 'Face Oil':4,'Serum':5,'Moisturizer':6, 'Primer':7, 'CC Cream': 8, 'BB Cream':9,
                 'Foundation':10, 'Concealer':11, 'Contour':12, 'Blush':13, 'Highlighter':14, 'Bronzer':15, 'Eye Shadow':16, 'Eyeliner':17, 'Mascara':18, 'Lip Liner':19,
                 'Lipstick':20, 'Lip Gloss':21, 'Powder':22, 'Setting Spray':23 }
# Apply map
df['Category_encoded'] = df['Category'].map(order_category)
df[['Category', 'Category_encoded']].head()
```

##### *Transformacion - Importando libreria preprocessing de sklearn*

```python
from sklearn.preprocessing import LabelEncoder

# Object LabelEncoder
pack_label_encoder = LabelEncoder()

# Encoding
df['Packaging_Type_encoded'] = pack_label_encoder.fit_transform(df['Packaging_Type'])

df[['Packaging_Type','Packaging_Type_encoded']].head()
```

##### **Transformacion de columnas categoricas NOMINALES**

```python
# One-Hot Encoding
brand_encoded = pd.get_dummies(df['Brand'], prefix='Brand')
gender_encoded = pd.get_dummies(df['Gender_Target'], prefix='Gender_Target')
df_new = pd.concat([df,brand_encoded, gender_encoded], axis=1)
df_new.head()
```

##### Columna Main_Ingredient

```python
main_ing_encoded = pd.get_dummies(df['Main_Ingredient'], prefix='Main_Ingredient')
df_new_main = pd.concat([df,main_ing_encoded], axis=1)
df_new_main.head()
```

```python
df.head()
```

```python
gender_encoded.head()
```

```python
# Add gender_encoded to df origin
df = pd.concat([df, gender_encoded,main_ing_encoded], axis=1)
df.head()
```

#### Normalización 


```python
df.head()
```

##### *Extración de informacion*
A la serie Product_Size, se le extraera el texto 'ml'.

```python
# Remove 'ml'
df['Product_Size_NoML'] = df['Product_Size'].str.extract(r'(\d+)ml', expand=False)
# Convert type from object to float
df['Product_Size_NoML'] = df['Product_Size_NoML'].astype(float)
df['Product_Size_NoML'].dtype
```

```python
df['Product_Size_NoML'].describe()
```

##### *Eliminación de espacios extras*


```python
# Delete space at the beginning, end and between
df.head()
for s in df.select_dtypes(include=['object']):
    df[s] = df[s].str.strip()
    df[s] = df[s].str.replace(r'\s+', ' ', regex=True)
```
